```python
from pathlib import Path 

import pandas as pd
```

```python
d = Path('./data/clustering')
episode_dirs = [x for x in d.iterdir()]
```

```python
data = []
for episode_dir in episode_dirs:
    cluster_dirs = [x for x in episode_dir.iterdir()]
    for cluster_dir in cluster_dirs:
        cluster = cluster_dir.parts[-1]
        files = [x for x in cluster_dir.iterdir()]
        for file in files:
            temp, frame_num, face_num = file.stem.split('_')
            datum = {'filename': file.name,
                     'season': int(temp[1:3]),
                     'episode': int(temp[4:6]),
                     'frame_num': int(frame_num),
                     'face_num': int(face_num),
                     'cluster': cluster}
            data.append(datum)
df = pd.DataFrame(data)
df.sort_values(by=['season', 'episode', 'frame_num', 'face_num'])
```

```python
faces_df = pd.read_csv('./data/faces.csv', index_col=0)
faces_df.head()
```

```python
cluster_df = faces_df.merge(df.drop('filename', axis=1),
                            on=['season', 'episode', 'frame_num', 'face_num'],
                            how='left')
cluster_df.head()
```
