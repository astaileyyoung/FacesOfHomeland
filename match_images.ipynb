```python
import os
import json
from pathlib import Path

import pandas as pd
```

```python
with open(r'.\data\characters.json', 'r') as f:
    characters = json.load(f)
characters
```

```python
images = []
for d in Path('.\data\characters').iterdir():
    files = [x for x in d.iterdir()]
    images.extend(files)
print(images[0])
```

```python
data = []
for image in images:
    name = image.stem
    character = image.parent.parts[-1].replace('_', ' ')
    temp, frame_num, face_num = name.split('_')
    try:
        datum = {
            'filename': image.name,
            'season': int(name[1:3]),
            'episode': int(name[4:6]),
            'frame_num': int(frame_num),
            'face_num': int(face_num),
            'character': character,
            'cast_id': str(characters[character])
        }
    except KeyError:
        print(image)
    data.append(datum)
df = pd.DataFrame(data)
df.head()
```

```python
faces_df = pd.read_csv(r'.\data\faces.csv', index_col=0)
faces_df.head()
```

```python
total_df = faces_df.merge(df,
                          on=['season', 'episode', 'frame_num', 'face_num'],
                          how='left')
total_df.head()
```

```python
total_df.to_csv(r'.\data\faces.csv')
```
