```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from imdb import Cinemagoer
from tqdm import tqdm
```

```python
url = 'https://www.imdb.com/search/title/?series=tt1796960&sort=user_rating,desc&count=250&view=simple'
page = requests.get(url)
content = BeautifulSoup(page.content, 'lxml')
spans = content.find_all('span', class_='lister-item-header')
```

```python
imdb_ids = []
for span in spans:
    link = span.find_all('a')[-1]['href']
    imdb_id = link[9:-1]
    imdb_ids.append(imdb_id)
```

```python
data = []
ia = Cinemagoer(loggingLevel=50)
for imdb_id in tqdm(imdb_ids):
    episode = ia.get_episode(imdb_id)
    datum = {'imdb_id': imdb_id,
             'title': episode.data['title'],
             'year': episode.data['year'],
             'season': episode.data['season'],
             'episode': episode.data['episode'],
             'cast': ','.join([x.personID for x in episode.data['cast']])}
    data.append(datum)
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('./data/episodes.csv')
```
