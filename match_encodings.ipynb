```python
from pathlib import Path 

import numpy as np
import pandas as pd
from tqdm.auto import tqdm
```

```python
df = pd.read_csv('./data/faces.csv', index_col=0)
df.head()
```

```python
df.iloc[0]['encoding_deepface']
```

```python
files = [x for x in Path('./data/encodings_deepface').iterdir()]
encodings = [np.load(str(x)) for x in tqdm(files)]
encodings[0]
```

```python
data = [{'filename': f'{x.stem}.png', 'encoding_deepface': str(list(encodings[num]))} for num, x in enumerate(tqdm(files))]
encoding_df = pd.DataFrame(data)
encoding_df.head()
```

```python
total_df = df.drop('encoding_deepface', axis=1).merge(encoding_df,
                    how='left',
                    on='filename')
total_df = total_df[['imdb_id',
                     'frame_num',
                     'face_num',
                     'x1',
                     'y1', 
                     'x2',
                     'y2',
                     'img_height', 
                     'img_width', 
                     'area',
                     'pct_of_frame',
                     'season',
                     'episode',
                     'encoding',
                     'encoding_deepface',
                     'filename',
                     'character',
                     'cast_id',
                     'cluster']]
total_df.head()
```

```python
total_df.to_csv('./data/faces.csv')
```
